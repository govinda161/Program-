```python
# Import the necessary libraries
import matplotlib.pyplot as plt  # For creating visualizations

# Sample data for the bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']  # Names of each category
values = [23, 45, 12, 36]  # Values corresponding to each category

# Create the bar chart
plt.bar(categories, values, color='orange')  # Create a bar chart with specified colors

# Adding labels and title
plt.xlabel('Categories')  # Set the label for the x-axis
plt.ylabel('Values')  # Set the label for the y-axis
plt.title('Bar Chart Example')  # Set the title of the chart

# Show the plot
plt.grid(axis='y')  # Add a horizontal grid for better readability
plt.show()  # Render the plot and display it to the user
```
