```python
# Import the pyplot module from matplotlib
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1, 2, 3, 4, 5]  # X-axis data points
y = [2, 3, 5, 7, 11]  # Y-axis data points

plt.scatter(x, y)  # Create a scatter plot with x and y data

plt.title('Simple Scatter Plot')  # Set the title of the plot
plt.xlabel('X-axis Label')  # Label for the X-axis
plt.ylabel('Y-axis Label')  # Label for the Y-axis

plt.grid(True)  # Enable the grid for better visibility
plt.show()  # Display the plot
```
